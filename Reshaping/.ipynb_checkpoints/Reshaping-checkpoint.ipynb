```python
import numpy as np
import pandas as pd
```

```python
med_data = pd.read_csv('../../data/claims_med_interns_2016-2018_20190304.csv')
med_data.columns
```

```python
def reshape(group):
    df = pd.DataFrame()
    df['gender'] = group.head(1)['Member Gender']
    df['age'] = group.head(1)['Age']
    df['postal_code'] = group.head(1)['Zip (5-digit)']
    df['county'] = group.head(1)['County']
    hypertension = ['I10' in s for s in group['Primary ICD Diagnosis Code']]
    df['binary_hypertension'] = True in hypertension
    cardiovascular = ['heart' in s for s in group['Primary ICD Rollup']]
    df['binary_cardiovascular'] = True in cardiovascular
    diabetes = ['Diabetes' in s for s in group['Primary ICD Rollup']]
    df['binary_diabetes'] = True in diabetes
    obesity = ['E66' in s for s in group['Primary ICD Diagnosis Code']]
    df['binary_obesity'] = True in obesity
    cancer = ['Cancer' in s for s in group['Primary ICD Rollup']]
    df['binary_cancer'] = True in cancer
    musculoskeletal = ['musculoskeletal' in s for s in group['Primary ICD Rollup']]
    df['binary_musculoskeletal'] = True in musculoskeletal
    copd = ['Chronic obstructive pulmonary disease' in s for s in group['Primary ICD Rollup']]
    df['binary_copd'] = True in copd
    cholesterol = ['E78' in s for s in group['Primary ICD Diagnosis Code']]
    df['binary_cholesterol'] = True in cholesterol
    
    df['claims_hypertension'] = sum(hypertension)
    df['claims_cardiovascular'] = sum(cardiovascular)
    df['claims_diabetes'] = sum(diabetes)
    df['claims_obesity'] = sum(obesity)
    df['claims_cancer'] = sum(cancer)
    df['claims_musculoskeletal'] = sum(musculoskeletal)
    df['claims_copd'] = sum(copd)
    df['claims_cholesterol'] = sum(cholesterol)
    return df
```

```python
reshaped_med_data = med_data.groupby(['Member ID Encrypted', 'Incurred Year']).apply(reshape)
```
